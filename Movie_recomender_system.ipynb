```python
import ast
ast.literal_eval('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
convert('[{"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 28, "name": "Action"}]')
```

```python
movies_credit["genres"]=movies_credit["genres"].apply(convert)
```

```python
def convert_top_3_cast(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i["name"])
            counter+=1
    return l
```

```python
movies_credit["cast"]=movies_credit["cast"].apply(convert_top_3_cast)
```

```python
movies_credit["crew"][1]
```

```python
def fetch_director(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            l.append(i["name"])
    return l
```

```python
movies_credit["crew"]=movies_credit["crew"].apply(fetch_director)
```

```python
movies_credit["overview"]=movies_credit["overview"].apply(lambda x:x.split())
```

```python
movies_credit["genres"]=movies_credit["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies_credit["keywords"]=movies_credit["keywords"].apply(lambda x:[i.replace(" ","") for i in x])

movies_credit["cast"]=movies_credit["cast"].apply(lambda x:[i.replace(" ","") for i in x])

movies_credit["crew"]=movies_credit["crew"].apply(lambda x:[i.replace(" ","") for i in x])

```

```python
movies_credit["tags"]=movies_credit["overview"]+movies_credit["genres"]+movies_credit['keywords']+movies_credit['cast']+movies_credit["crew"]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem("loved")
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. Action Adventure Fantasy ScienceFiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors=cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors[1]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[14])),reverse=True,key= lambda x:x[1])[1:6]
```

```python
sorted(similarity[1],reverse=True)
```

```python
def recommend(movie):
    movie_index= new_df[new_df["title"]==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key= lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend("Avatar")
```
